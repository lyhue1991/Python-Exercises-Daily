给定字符串abc##dfg##gh，实现将所有#移至字符串串头。输出####abcdfggh。

```python
#==============================提示区==============================#




































































#  提示：可以用列表操作或者使用正则表达式
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：正则表达式

s = "abc##dfg##gh"

import re
def f(s):
    result = re.sub("[^#]","",s)+re.sub("#","",s)
    return(result)
f(s)

#==============================答案区==============================#
```
