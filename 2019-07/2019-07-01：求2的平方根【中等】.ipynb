# 求2的平方根，保留小数点后10位【中等】

求2的平方根，保留小数点后10位。

提示：不可调用系统sqrt函数。

```python
#==============================提示区==============================#




































































# 提示：二分法
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路，二分查找

n = 10
def f(n):
    low,high = 1,2
    while high - low >= 1*10**(-n-1):
        mid = (low + high)/2.0
        if mid**2 < 2:
            low = mid
        else:
            high = mid
    print("sqrt(2) = {:.{}f}".format(mid,n))
    return(round(mid,n))

f(10)

#==============================答案区==============================#
```
