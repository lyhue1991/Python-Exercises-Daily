```python
#==============================答案区==============================#

# 解题思路：哈希查找+动态规划

s = "abcbefgf"
def f(s):    
    dic = {}
    start,maxlen,substr = 0,0,""

    for i,x in enumerate(s):
        if x in dic:
            start = max(dic[x]+1,start)
            dic[x] = i   
        else:
            dic[x] = i

        if i-start+1>maxlen:
            maxlen = i-start+1
            substr = s[start:i+1]
    return(substr)

print(f("abcbefgf"))
print(f("abcdef"))
print(f("abbcddefh"))
#==============================答案区==============================#
```
