# 跳跃游戏【中等】

跳跃游戏

给定一个非负整数数组，你最初位于数组的第一个位置。

数组中的每个元素代表你在该位置可以跳跃的最大长度。

判断你是否能够到达最后一个位置。

示例 1:

输入: [2,3,1,1,4]

输出: true

解释: 从位置 0 到 1 跳 1 步, 然后跳 3 步到达最后一个位置。

示例 2:

输入: [3,2,1,0,4]

输出: false

解释: 无论怎样，你总会到达索引为 3 的位置。

但该位置的最大跳跃长度是0，所以你永远不可能到达最后一个位置。

```python
#==============================提示区==============================#




































































#  提示：贪心算法，每次跳时选择index+value最大的位置，如果遇到了0则说明跳不出来。
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路1：存在某个0，且这个0前面的每个数的大小都小于等于它和0之间的距离，则说明跳不出来

def f(arr):
    zeros_idx = [i for i,x in enumerate(arr) if x==0]
    for z in zeros_idx:
        if all([arr[i]<= z-i for i in range(0,z)]):
            return(False)
    return(True)

print(f([1,2,3]))
print(f([3,2,1,0,1]))
print(f([4,2,0,2,0,1]))
#==============================答案区==============================#
```

```python
#==============================答案区==============================#
# 解题思路2：贪心算法
# 每次跳时选择index+value最大的位置，如果遇到了0则说明跳不出来了
# 每个位置i能往前跳的步数<=arr[i]

def f(arr):
    i = 0
    while i<len(arr)-1:
        if arr[i]==0:
            return(False)
        reachables = range(i+1,min(i+arr[i]+1,len(arr)))  
        i = sorted(reachables,key = lambda k: k+arr[k],reverse = True)[0]
    if i>=len(arr)-1:
        return(True)
print(f([1,2,3]))
print(f([3,2,1,0,1]))
print(f([4,2,0,2,0,1]))
#==============================答案区==============================#
```
