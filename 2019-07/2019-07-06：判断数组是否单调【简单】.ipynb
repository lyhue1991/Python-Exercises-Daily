# 判断数组是否单调【简单】

当给定的数组是单调数组时返回 true ，否则返回 false 。

单调可以是单调递增或者单调递减。

示例 1：

输入：[1,2,2,3]

输出：true

示例 2：

输入：[6,5,4,4]

输出：true

示例 3：

输入：[1,3,2]

输出：false

```python
#==============================答案区==============================#
# 解题思路：使用numpy数组实现向量化计算

import numpy as np

def f(arr):
    result = all(np.array(arr[1:]) >= np.array(arr[0:-1])) or \
             all(np.array(arr[1:]) <= np.array(arr[0:-1]))
    return(result)

print(f([1,2,3,3,4,5,6,6,7]))
print(f([3,3,3,2,2,1]))
print(f([1,2,3,2,1]))

#==============================答案区==============================#
```
