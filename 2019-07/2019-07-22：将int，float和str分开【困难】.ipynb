# 将int，str，和float分开【简单】

给定一个列表，其中由整数和浮点数和字符串。

将int放在列表前面位置，float放在列表后面位置，将str放在中间位置。

在原列表上操作。

例如：arr = [1,“hello”,0.1,3,0.4,5,"world"]

输出为

 [1,5,3,“hello”,"world",0.4,0.1]

```python
#==============================提示区==============================#




































































#  提示：可以通过交换的方式，从首尾同时操作
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：首尾各用一个变量记录下一个int和float要放置的位置
# 扫描数组，如果是int就交换到前面，如果是float就交换到后面

arr = [1,"hello",0.1,3,0.4,5,"world"]

def f(arr):
    if len(arr)<=1:
        return(arr)
    i,j = 0,len(arr)-1
    k = 0
    while k<=j:
        if isinstance(arr[k],int):
            arr[i],arr[k] = arr[k],arr[i]
            i = i+1
            k = max(k,i)
        elif isinstance(arr[k],float):
            arr[j],arr[k] = arr[k],arr[j]
            j = j-1
        else:
            k = k+1
        print(arr,"i=%d,j=%d,k=%d"%(i,j,k))
    return(arr)
#==============================答案区==============================#
```

```python
f(arr)
```
