# 递增数组中是否存在a+b=c【简单】

判断递增数组中是否存在3个数a,b,c，满足a+b=c

例1：

arr = [1,2,3,4],

存在 a = 1,b = 2,c = 3,满足a+b=c

输出 True

例2：

arr = [1,5,9,18],

不存在满足条件的a,b,c

输出 False

```python
#==============================提示区==============================#




































































# 提示：3重遍历
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：3重遍历

arr = [1,2,3,4]

def f(arr):
    n = len(arr)
    for i in range(0,n-2):
        for j in range(i+1,n-1):
            for k in range(j+1,n):
                if arr[k] == arr[i] + arr[j]:
                    return(True)
    return(False)

f(arr)

#==============================答案区==============================#
```
