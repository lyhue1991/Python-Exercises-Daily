```python
#==============================答案区==============================#
#解题思路：蒙特卡洛算法
#采样0到1之间的两个随机数，作为x,和y，
#如果（x-0.5）**2+（y-0.5）**2<=1/4则说明它们落在圆内。
#重复这个试验10000次，统计落在圆内的点数量。用这个数量就可以估计pi.

import numpy as np 
n = 100000
points = np.random.uniform(0,1,size = (n,2))
xs,ys = points.reshape(2,-1)
pi = np.sum((xs - 0.5)**2 + (ys - 0.5)**2 <=1/4)/n*4
pi
#==============================答案区==============================#
```
