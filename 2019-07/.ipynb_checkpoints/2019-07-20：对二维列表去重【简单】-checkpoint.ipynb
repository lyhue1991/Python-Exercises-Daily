# 数组全排列【中等】

给定一个数组，找出其所有可能的排列。

例如:

arr = [1,1,3]

输出为

[[1,1,3],[1,3,1],[3,1,1]]


```python
#==============================答题区==============================#




















#==============================答题区==============================#
```

```python
#==============================提示区==============================#




































































#  提示：递推思想
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：使用递推方法

import numpy as np 
arr = [1,1,2,3]

def f(arr):
    if len(arr)<=1:
        return([arr])
    t = [arr[0:1]]
    i = 1
    while i<=len(arr)-1:
        a = arr[i]
        t = [xs[0:j]+[a]+xs[j:] for xs in t for j in range(i+1)]
        t = np.unique(t,axis=0).tolist()
        i = i+1
    return(t)

f(arr)           
#==============================答案区==============================#
```
