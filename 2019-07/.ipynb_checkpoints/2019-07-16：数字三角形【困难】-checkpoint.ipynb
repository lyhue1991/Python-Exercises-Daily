使用单线程CPU，最多可以跑这些任务中的多少个任务？注：因为是单线程CPU，任务之间不能有时间重叠。

```python
#==============================答案区==============================#
# 解题思路：动态规划算法
# 先对所有任务按照 开始时间排序，从头开始扫描，
# 记录两个状态，即任务结束时间end_dt和任务数量n，
# 如果下一个任务的开始时间 > end_dt, 任务数量加1，任务结束时间对应增加
# 如果下一个任务的结束时间 < end_dt，用这个任务替换已有任务序列中最后一个任务
# 任务数量不变，但end_dt对应变短。

tasks = [(1,4),(2,3),(4,9),(5,7),(8,12)]

def f(tasks):
    if len(tasks) <=1:
        return(len(tasks))
    tasks = sorted(tasks,key = lambda x:x[0])
    start,end = tasks[0]
    n = 1
    for t in tasks[1:]:
        if t[0]>end:
            n = n+1
            end = t[1]
        else:
            if t[1]<end:
                end = t[1]
    return(n)
f(tasks)
#==============================答案区==============================#
```
