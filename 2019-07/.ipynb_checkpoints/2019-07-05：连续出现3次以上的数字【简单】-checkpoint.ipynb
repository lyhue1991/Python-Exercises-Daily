# 找出数组中连续出现3次以上的数字【简单】

给定一个列表，找出其中连续出现3次以上的数字。

例如，

输入，arr=[1,3,3,1,2,2,2,2,5,5,5,3]。

输出，［2，5］。

```python
#==============================提示区==============================#




































































# 提示：通过索引遍历数组
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：使用numpy数组实现向量化计算

arr = [1,2,2,2,3,3,1,1,5,5,5,5,1]

def f(arr):
    arr1 = arr[0:-2]
    arr2 = arr[1:-1]
    arr3 = arr[2:]
    
    result = list(set([arr1[i] for i in range(len(arr1)) 
    if arr1[i]==arr2[i] and arr1[i]==arr3[i]]))
    return(result)

f(arr)
#==============================答案区==============================#
```
