# 字符串按照单词反转【简单】

将字符串按照单词翻转。

例如：输入s = "Python love I"
输出为 "I love Python"

```python
#==============================提示区==============================#




































































#  提示：字符串常用函数split,join,reverse
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：先以空格分隔为列表，reverse后用空格join

s = "Python Love I"

def f(s):
    result = " ".join(s.split(" ")[::-1])
    return(result)
f(s)
#==============================答案区==============================#
```
