# wordcount词频统计【简单】

wordcount词频统计

给定一个文档doc.txt，内容如下

hello world

hello China

hello BeiJing

读取其中的内容，并统计每个单词出现的次数。

这个例子中的输出应当为

hello：3

world：1

China：1

BeiJing：1

单词顺序不重要。

```python
#==============================提示区==============================#




































































#  提示：读文件，字典
































































































































#==============================提示区==============================#
```

```python
%%writefile doc.txt
hello world
hello China
hello BeiJing
```

```python
#==============================答案区==============================#
# 解题思路：用字典统计词频

def wordcount(filename):
    dic = {}
    with open(filename) as f:
        for line in f.readlines():
            words = line[:-1].split(" ")
            for w in words:
                dic[w] = dic.get(w,0)+1 
    for word,count in dic.items():
        print(word,':',count)
    return(dic)

wordcount("doc.txt")    
#==============================答案区==============================#
```
