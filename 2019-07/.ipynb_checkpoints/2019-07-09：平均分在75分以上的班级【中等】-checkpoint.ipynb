找出平均分在75分以上的班级

已知班级信息表和成绩表，找出班级平均分在75分以上的班级
班级信息表包括cls,name,成绩表包括name,score

```python
#==============================答案区==============================#
# 解题思路：先merge连接，再按照班级groupby

classes = [("class1","LiLei"),
          ("class1","HanMeiMei"),
          ("class2","DaChui"),
          ("class2","RuHua")]
          
scores = [("LiLei",76),
          ("HanMeiMei",80),
          ("DaChui",70),
          ("RuHua",60)]

import numpy as np 
import pandas as pd
def f(classes,scores):
    dfclass = pd.DataFrame(classes,columns = ["class","name"])
    dfscore = pd.DataFrame(scores,columns = ["name","score"])
    result = dfclass.merge(dfscore,on = ["name"]).groupby("class").mean().query("score>75")
    return(result.index.values.tolist())

f(classes,scores)
#==============================答案区==============================#
```
