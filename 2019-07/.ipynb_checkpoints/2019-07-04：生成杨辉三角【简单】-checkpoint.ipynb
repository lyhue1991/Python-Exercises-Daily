# 生成杨辉三角【简单】

给定一个正整数n， 生成杨辉三角的前n行。

输入: 5

输出:

[

[1],

[1,1],

[1,2,1],

[1,3,3,1],

[1,4,6,4,1]

]

```python
#==============================答题区==============================#




















#==============================答题区==============================#
```

```python
#==============================提示区==============================#




































































# 提示：杨辉三角的生成公式是每一行的首尾都是1，其余元素是其两肩膀元素之和。
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：利用递归


def f(n):
    if n==1:
        return([[1]])
    elif n==2:
        return([[1],[1,1]])
    else:
        p = f(n-1)
        x = p[-1]
        result = p + [[1] + [x[i] + x[i+1] for i in range(len(x)-1)] + [1]]
        return(result)
f(7)
#==============================答案区==============================#
```
