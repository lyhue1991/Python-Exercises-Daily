# 用Python实现堆排序算法【困难】

堆排序算法
1，堆排序先自下而上构造最大堆（从小到大排序）
2，然后交换堆顶和堆底，下沉最大值，堆的大小减小1
3，将其余部分重新构造成一个最大堆
4，不断重复2和3步骤，直到将数组变得从小到大排序

```python
#==============================提示区==============================#




































































# 提示：最大堆满足 arr[i] >= arr[2*i+1] and arr[i]>=arr[2*i+2]
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：最大堆满足 arr[i] >= arr[2*i+1] and arr[i]>=arr[2*i+2]


# 对不满足最大堆条件的节点进行交换，i为节点序号，n为堆的大小
def swap_node(arr,i,n):
    if 2*i+2<=n-1 and arr[2*i+2] > max(arr[2*i+1],arr[i]):
        arr[i],arr[2*i+2] = arr[2*i+2],arr[i]
        swap_node(arr,2*i+2,n)
    elif 2*i+1<=n-1 and arr[2*i+1] > arr[i]:
        arr[i],arr[2*i+1] = arr[2*i+1],arr[i]
        swap_node(arr,2*i+1,n)
    else:
        pass
    
def heap_sort(arr):
    # 初始化堆
    for i in range(len(arr)//2-1,-1,-1):
        swap_node(arr,i,len(arr))
    # 排序过程
    for n in range(len(arr),1,-1):
        # 交换顶底
        arr[0],arr[n-1] = arr[n-1],arr[0]
        # 从顶开始重建新的堆
        swap_node(arr,0,n-1)
    return(arr)

heap_sort([1,1,1,3,3,3,2,2,2,6,6,6,5,5,5])

#==============================答案区==============================#
```
