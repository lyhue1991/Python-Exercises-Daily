# 二维列表去重【简单】

给定一个二维列表，对其去重。

例如:

arr = [[1,3],[1,1],[1,3],[1,2]]

输出为

[[1,3],[1,1],[1,2]]

```python
#==============================提示区==============================#




































































#  提示：可以使用np.unique或者集合
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路1：使用集合去重

arr = [[1,3],[1,1],[1,3],[1,2]]

def f(arr):
    result = list([list(x) for x in set([tuple(x) for x in arr])])
    return(result)

f(arr)
#==============================答案区==============================#
```

```python
#==============================答案区==============================#
# 解题思路2：使用numpy.unique去重

arr = [[1,3],[1,1],[1,3],[1,2]]
import numpy as np 
def f(arr):
    result = np.unique(arr,axis = 0).tolist()
    return(result)

f(arr)
#==============================答案区==============================#
```
