# 判断两个字符串是否同构【中等】

判断字符串s是否由词典wordDict中的词构成，词典中的词可以重复使用，也可以有些词不用。

例1：s = "helloworld"，wordDict = ["hello","word","world"] ，答案是True

例2：s = "helloworldhelloChina"，wordDict = ["hello","China","world"] ，答案是True

例3：s = "helloworldhelloChina"，wordDict = ["hello","word","world"] ，答案是False

```python
#==============================答案区==============================#
# 解题思路：回溯法
# 尝试从wordDict中寻找和s的开头部分匹配的词，如果找到，将s砍去匹配部分得到新的s。
# 继续在wordDict中寻找和新的s的开头部分匹配的词，直到s的长度变为0，返回True
# 如果在当中任何一个步骤中失败了，返回上一步，寻找别的词进行匹配。
# 如果尝试完所有可能后s的长度都没有变成0，返回False

def f(s,wordDict):
    d = [(0,wordDict[0])]
    while True:
        i,w = d[-1]
        if s.startswith(w):
            s = s[len(w):]
            if len(s)==0:
                return(True)
            else:
                d.append((0,wordDict[0]))
        else:
            if all([x[0]==len(wordDict)-1 for x in d]):
                return(False)
            if i==len(wordDict)-1:
                d.pop()
                s = d[-1][-1]+s
            i,w = d[-1]
            i,w = i+1,wordDict[i+1]
            d[-1] = (i,w)        
            

print(f("helloworld",["he","hello","world","word"]))
print(f("helloworldhelloChina",["hello","world","China"]))
print(f("helloworldhelloChina",["hello","world","word"]))     
#==============================答案区==============================#
```
