# 将int和float分开【简单】

给定一个列表，由整数和浮点数组成。

将int放在列表前面，float放在列表后面位置，从而将两者分开。

在原列表上操作。

例如：arr = [1,2,2.3,4,3.2,5]

输出为

[1,2,4,5,2.3,3.2]

```python
#==============================提示区==============================#




































































#  提示：可以通过交换的方式操作
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#
# 解题思路：首尾各用一个变量记录下一个int和float要放置的位置
# 扫描数组，如果是int就前进，如果是float就交换到后面

arr = [1,2,2.3,4,3.2,5]

def f(arr):
    if len(arr)<=1:
        return(arr)
    i,j = 0,len(arr)-1
    while i<j:
        if isinstance(arr[i],int):
            i = i+1
        else:
            arr[i],arr[j] = arr[j],arr[i]
            j = j-1
    return(arr)
#==============================答案区==============================#
```

```python
f(arr)
```
