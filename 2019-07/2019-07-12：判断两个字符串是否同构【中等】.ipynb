# 判断两个字符串是否同构【中等】

两个字符串同构指的是它们的组成字符存在一一对应关系。

如果有多对一或者一对多关系，则不同构。

例1 ：s=“egg”，t=“add”，输出 True

例2：s=“egg”，t=“abc”，输出 False

例3：s=“egg”，t=“aaa”，输出 False

```python
#==============================答案区==============================#
# 解题思路：哈希表。
# 同步扫描两个字符串，用2个dict分别记录s->t和t->s的字符对应关系
# 如果扫描过程中出现了一对多或者多对一冲突，输出False
# 扫描完毕后未出现冲突，输出True.


import pdb
def f(s,t):
    if len(s)!=len(t):
        return(False)
    s2t,t2s = {},{} 
    for x,y in zip(s,t):
        if x in s2t :
            if s2t[x]!=y:
                return(False)
        elif y in t2s:
            if t2s[y]!=x:
                return(False)
        else:
            s2t[x] = y
            t2s[y] = x
    else:
        return(True)
print(f("egg","add"))
print(f("egg","abc"))
print(f("egg","aaa"))       
#==============================答案区==============================#
```
