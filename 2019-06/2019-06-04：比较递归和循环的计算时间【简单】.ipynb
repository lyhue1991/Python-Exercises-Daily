# 比较递归和循环的计算时间【容易】

斐波那契数列满足：

fib(1)=fib(2)=1

fib(n) = fib(n-1)+fib(n-2) 

下面为两种计算斐波那契数列的方法。比较它们的计算时间，哪种方法更快一些？


```python
# 求斐波那契数列第30个元素

### 方法A:利用循环

def fib(n):
    if n in {1,2}:
        return(1)
    i = 2
    a,b = 1,1
    while i<n:
        i+=1
        a,b=b,a+b
    return(b)
print(fib(30))

### 方法B:利用递归

def fib(n):
    return(1 if n in (1,2) else fib(n-1)+fib(n-2))
print(fib(30))
```

```python
#==============================答案区==============================#

# 求斐波那契数列第30个元素

### 方法A:利用循环

def fib(n):
    if n in {1,2}:
        return(1)
    i = 2
    a,b = 1,1
    while i<n:
        i+=1
        a,b=b,a+b
    return(b)
%time fib(30)

### 方法B:利用递归

def fib(n):
    return(1 if n in (1,2) else fib(n-1)+fib(n-2))
%time fib(30)



#递归的实现涉及到使用堆栈保存现场，在这个例子中会导致大量的重复计算。
#递归虽然可以增加代码的可读性，但其效率通常比较低。
#程序中应当避免过深的递归。

#==============================答案区==============================#
```
