```python
#==============================答案区==============================#

# 解题思路：到达第n级台阶有两种可能，要么是从第n-1级台阶走一步来的，
# 要么是从n-2级台阶走两步来的，所以f(n)= f(n-1)+f(n-2)

def f(n):
    if n==1:
        return(1)
    elif n==2:
        return(2)
    else:
        return(f(n-1)+f(n-2))
    
f(10)

#==============================答案区==============================#
```
