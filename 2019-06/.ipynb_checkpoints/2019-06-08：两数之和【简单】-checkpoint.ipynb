# 压平列表【简单】

给定一个二维的列表，即这个列表的元素也是一些列表，如何把它们压平成一个一维的列表？

例如：输入为[[1,2,3],["a","b"],[5,"c"]]

输出应当是[1,2,3,"a","b",5,"c"]

```python
#==============================答题区==============================#




















#==============================答题区==============================#
```

```python
#==============================提示区==============================#




































































#  提示：二重列表推导式
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#

# 解题思路：用列表推导式

arr = [[1,2,3],["a","b"],[5,"c"]]

def f(arr):
    return([x for l in arr for x in l])

f(arr)

# [x for l in arr for x in l] 等价成以下循环
#
# output = []
# for l in arr:
#    for x in l:
#        output.append(x)

#==============================答案区==============================#
```
