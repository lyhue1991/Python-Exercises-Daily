# 比较in查找的计算时间

数组和集合查找某个元素所需时间，谁更快呢？

具体地说以下是A方式更快还是B方式更快呢？

```python
data = [i**2 + 1 for i in range(1000000)]
list_data = list(data)
set_data = set(data)


#A方式查找
1098987 in list_data

#B方式查找
1098987 in set_data
```
