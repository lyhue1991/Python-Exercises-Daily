```python
#==============================答案区==============================#

# 解题思路：使用extend合并列表，使用append追加元素到列表。递归作用

arr = [1,2,[3,5],[4,6,[7,8]]]
def f(arr):
    result = []
    for x in arr:
        if isinstance(x,list):
            result.extend(f(x))
        else:
            result.append(x)
    
    return(result[0] if result else [])

f(arr,target)

#==============================答案区==============================#
```
