# 两数之和【简单】

寻找列表中满足两数之和等于目标值的元素的下标。

例如：arr = [2,7,4,9],target = 6

则返回 [0,2],若不存在，返回空列表[].

```python
#==============================提示区==============================#




































































#  提示：哈希表
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#

# 解题思路：使用列表推导式二重循环遍历，复杂度O(n**2)

arr = [2,7,4,9]
target = 6
def f(arr,target):
    result = [(i,j) for i in range(len(arr)) 
              for j in range(len(arr)) 
             if i<j and arr[i]+arr[j]==target]
    
    return(result[0] if result else [])

f(arr,target)

#==============================答案区==============================#
```

```python
#==============================答案区==============================#

# 解题思路：用字典结构，哈希查找减少一重遍历，复杂度 O(n)


arr = [2,7,4,9]
target = 6

def f(arr,target):
    dic = {}
    for i,x in enumerate(arr):
        j = dic.get(target-x,-1)
        if j != -1:
            return((j,i))
        else:
            dic[x] = i
    return([])

f(arr,target)

#==============================答案区==============================#
```
