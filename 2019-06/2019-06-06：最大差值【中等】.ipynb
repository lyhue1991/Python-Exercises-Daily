# 最大差值【中等】

有一个列表，例如[10,4,2,6,4,3,9,8,5],

求其中两个数x,y，

满足x的索引小于y的索引，使得 y-x 最大。

(这个示例的答案是x=2,y=9,最大值为7)

```python
#==============================提示区==============================#




































































#  提示：动态规划
































































































































#==============================提示区==============================#
```

```python
#==============================答案区==============================#

# 解题思路1：穷举法，n**2复杂度，效率偏低

arr = [10,4,2,6,4,3,9,8,5]

def f(arr):
    assert len(arr)>2, "len(arr) should > 2!"
    pairs = [(x,y) for i,x in enumerate(arr) 
             for j,y in enumerate(arr) if i<j]
    diffs = [y-x for x,y in pairs]
    maxdiff = max(diffs)
    x,y = pairs[diffs.index(maxdiff)]
    print("x=",x,",y=",y)
    
    return(x,y)

f(arr)

#==============================答案区==============================#
```

```python
#==============================答案区==============================#

# 解题思路2：动态规划法，只用扫描一遍，效率较高。
# 动态规划的思想跟递归有些相似，用一些状态记录第k-1步时和目标相关的变量。
# 如果在第k步时能够用合理的策略更新这些量，就可以往前推进。


# 这里设置4个状态量：xmin记录之前扫描过的数当中的最小值,
# maxdiff记录扫描到的满足条件的最大差值,
# x,y记录取maxdiff时对应的x和y


arr = [10,4,2,6,4,3,9,8,5]

def f(arr):
    assert len(arr)>2, "len(arr) should > 2!"
    x,y = arr[0:2]
    xmin = x
    maxdiff = y-x
    
    for i in range(2,len(arr)):
        if arr[i-1] < xmin:
            xmin = arr[i-1]
        if arr[i] - xmin > maxdiff:
            maxdiff = arr[i] - xmin
            x,y = xmin,arr[i]
            
    print("x=",x,",y=",y)
    return(x,y)

f(arr)

#==============================答案区==============================#
```
